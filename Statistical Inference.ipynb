```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def poisson_distribution(n):
    # Simulate a sample from the Poisson distribution
    sample = np.random.poisson(5, n)

    # Plot the histogram of the sample
    plt.hist(sample, bins=np.arange(0, 20), alpha=0.6, color='g', label='Sample Histogram')
    plt.legend()
    plt.show()

    # Plotting the Q-Q plot comparing the sample to the theoretical Poisson distribution
    stats.probplot(sample, dist=stats.poisson, sparams=(5,), plot=plt)
    plt.show()

# Simulating and plotting the distribution for n = 10
poisson_distribution(10)

# Simulating and plotting the distribution for n = 100
poisson_distribution(100)

# Simulating and plotting the distribution for n = 1000
poisson_distribution(1000)
```

```python

def poisson_distribution(n, sample_size):
    means = []
    for i in range(sample_size):
        # Simulating a sample from the Poisson distribution
        sample = np.random.poisson(5, n)

        # Computing the empirical mean of the sample
        mean = np.mean(sample)
        means.append(mean)

    # Plotting the histogram of the 100 empirical means
    plt.hist(means, bins=20, alpha=0.6, color='g', label='Histogram of Empirical Means')
    plt.legend()
    plt.show()

# Simulating 100 samples of size 10
poisson_distribution(10, 100)
```

This result is known as the Central Limit Theorem, which states that the sum of a large number of independent random variables with finite mean and variance approaches a Normal distribution, regardless of the distribution of the individual random variables.

C)As the sample size n increases, the histogram of the 100 empirical means becomes more and more symmetric and bell-shaped, and the variance of the distribution decreases. This is due to the Central Limit Theorem.

The Central Limit Theorem states that the sum of a large number of independent random variables with finite mean and variance approaches a Normal distribution, regardless of the distribution of the individual random variables. As the sample size n increases, the sum of the random variables becomes more and more Normal-distributed, and the variance of the distribution decreases, approaching the population standard deviation divided by the square root of the sample size (np.sqrt(5/n) in this case).

In other words, as n increases, the empirical means become more and more tightly clustered around the population mean, and the histogram of the 100 empirical means becomes more and more Normal-distributed, with a smaller variance and a more pronounced peak. This is why you observe that the histogram becomes more symmetric and bell-shaped as n increases.

The distribution of F^(-1)(U) is the same as the distribution of the random variable X, where X = F^(-1)(U). This is because the inverse function F^(-1) transforms the uniform random variable U on (0, 1) into a random variable X on (a, b).

Since F is one-to-one and strictly increasing on (a, b), it follows that the inverse function F^(-1) is also one-to-one and strictly increasing on (0, 1). Therefore, for any set A in (a, b), the pre-image of A under F^(-1) has the same probability as the pre-image of F(A) under U, that is:

P(F^(-1)(U)  A) = P(U  F(A))

Since U is uniformly distributed on (0, 1), it follows that F^(-1)(U) has the same distribution as X, which is the random variable whose cdf is F^(-1).

B)One way to sample from the Exponential distribution Exp() using a sample from the uniform distribution U(0,1) is through the inverse transform method. Here's the algorithm:

Generate a sample U from the uniform distribution U(0, 1).

Calculate the inverse of the cumulative distribution function (CDF) of the Exponential distribution at U:
X = -(1/) * ln(1 - U)

where ln denotes the natural logarithm and  is the rate parameter of the Exponential distribution.

The value X is a sample from the Exponential distribution Exp().

This algorithm works because the CDF of the Exponential distribution is F(x) = 1 - e^(-x), and its inverse is F^(-1)(u) = -(1/) * ln(1 - u). By transforming a sample from the uniform distribution U(0,1) into the exponential distribution through this inverse function, we can generate samples from the exponential distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

def sample_from_exponential(lambda_, n_samples):
    u = np.random.uniform(0, 1, n_samples)
    x = -(1/lambda_) * np.log(1 - u)
    return x

# Generate 1000 samples from the Exponential distribution Exp() with  = 1
n_samples = 1000
lambda_ = 1
samples = sample_from_exponential(lambda_, n_samples)

# Plot the histogram of the generated samples
plt.hist(samples, bins=50, density=True, label='Generated samples')

# Plot the theoretical Exponential distribution Exp() with  = 1
x = np.linspace(0, 10, 100)
y = lambda_ * np.exp(-lambda_ * x)
plt.plot(x, y, 'r', label='Exponential distribution')

plt.legend()
plt.show()
```
